# Preparing Data 

```python
training_set.drop(['Names','emails','Country'], axis =1, inplace=True)
```

```python
training_set
```

```python
X =training_set.drop('Clicked', axis=1).values
```

```python
X
```

```python
X.shape
```

```python
type(X)
```

```python
y =training_set['Clicked'].values
```

```python
y
```

```python
y.shape
```

```python
type(y)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc =StandardScaler()
X =sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
classifier =LogisticRegression(random_state=10)
classifier.fit(X_train, y_train)
```

```python
y_train_predict =classifier.predict(X_train)
```

```python
y_train_predict
```

```python
from sklearn.metrics import confusion_matrix
cm =confusion_matrix(y_train, y_train_predict)
sns.heatmap(cm, annot=True, fmt='d')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train, y_train_predict))
```

```python
y_test_predict =classifier.predict(X_test)
```

```python
y_test_predict
```

```python
cm =confusion_matrix(y_test, y_test_predict)
sns.heatmap(cm, annot=True, fmt='d')
```

```python
print(classification_report(y_test, y_test_predict))
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set =X_train, y_train
X1, X2 =np.meshgrid(
    np.arange(start =X_set[:, 0].min() - 1, stop =X_set[:, 0].max() + 1, step=0.01 ),
    np.arange(start =X_set[:, 1].min() - 1, stop =X_set[:, 1].max() + 1, step=0.01)
)
plt.contourf(X1,X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T ).reshape(X1.shape),
            alpha=0.75, cmap=ListedColormap(('magenta','blue'))
            )
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set ==j, 0], X_set[y_set ==j, 1], c=ListedColormap(('magenta', 'blue'))(i), label=j )
plt.title('Facebook Add Customer Click Prediction (Training Set)')
plt.xlabel('Time Spent on Site')
plt.ylabel('Estimated Salry')
plt.legend()
plt.show()
```
